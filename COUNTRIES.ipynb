```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests

url = 'https://www.worldometers.info/world-population/population-by-country/'

# Get the HTML content
html = requests.get(url).content

# Parse the HTML content
soup = bs(html, 'html.parser')

# Find the table containing the population data
table = soup.find('table', {'id': 'example2'})

# Extract the table headers
headers = [th.text.strip() for th in table.find('thead').find_all('th')]

# Extract the table rows
rows = []
for tr in table.find('tbody').find_all('tr'):
    rows.append([td.text.strip() for td in tr.find_all('td')])

# Create a pandas dataframe
df = pd.DataFrame(rows, columns=headers)

# Print the dataframe
print(df)
```

```python
# Drop the specified columns
df.drop(columns=['Yearly Change', 'Net Change', 'Density (P/Km²)', 'Migrants (net)', 'Fert. Rate', 'Urban Pop %', 'World Share'], inplace=True)

# Print the updated dataframe
print(df.head())
```

```python
import random

# Get a random country from the 'Country (or dependency)' column
random_country = random.choice(df['Country (or dependency)'])

# Prompt the user to input the population of the random country
user_input = input(f"What is the population of {random_country} in 2023? ")

# Get the actual population of the random country from the dataframe
actual_population = df.loc[df['Country (or dependency)'] == random_country, 'Population (2023)'].values[0]

# Check if the user input matches the actual population
if user_input == actual_population:
    print("Correct!")
else:
    print(f"Incorrect. The population of {random_country} in 2023 is {actual_population}")
```
